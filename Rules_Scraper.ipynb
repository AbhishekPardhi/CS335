```python
from bs4 import BeautifulSoup
import requests

check = []
common_prefix = "https://docs.oracle.com/javase/specs/jls/se17/html/"
max_depth = 5

def PrintRules(url, depth):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    productions = soup.find_all('div', class_='production')

    results = []

    for production in productions:
        lhs = production.find('div', class_='lhs')
        rhs = production.find('div', class_='rhs')
        lhs_text = lhs.get_text().strip()[:-1]

        rule = lhs_text + ":\n\t"
        
        for element in rhs.contents:
            # If the element is a <br>, print " | "
            if element.name == 'br':
                rule += "\n|\t"
            # If the element is an <a> or <code>, print its text
            elif element.name == 'a':
                key = element.get_text().strip()
                rule += key+" "
                if(key not in check):
                    new_url = common_prefix+element['href']
                    check.append(new_url)
                    if(depth > 0):
                        PrintRules(new_url, depth-1)
            elif element.name == 'code':
                rule += element.get_text().strip()
            # If the element is a string, print it
            elif isinstance(element, str):
                rule += element.strip()+" "
                if(element.strip()=="}"):
                    rule += "*"
                elif(element.strip()=="]"):
                    rule += "?"
        
        # Print the final rule
        print(rule,"\n;\n\n")

open('output.txt', 'w') as file
url = 'https://docs.oracle.com/javase/specs/jls/se17/html/jls-19.html'
PrintRules(url, max_depth)
```
